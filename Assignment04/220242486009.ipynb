```python
#1.1 Load dataset and display first 10 transactions
```

```python
import pandas as pd

faceplate = pd.read_csv("Faceplate.csv")

faceplate.head(10)
```

```python
#1.2 Support of itemset {red, white}
```

```python
total_transactions = len(faceplate)

support_count = len(faceplate[(faceplate['Red'] == 1) &
                               (faceplate['White'] == 1)])

support = support_count / total_transactions

print("Support count:", support_count)
print("Support:", support)
```

```python
#2.1 Apply Apriori (min_support = 0.2)
```

```python
from mlxtend.frequent_patterns import apriori

faceplate_items = faceplate.drop(columns=['Transaction'])

frequent_itemsets = apriori(faceplate_items,
                            min_support=0.2,
                            use_colnames=True)

frequent_itemsets.sort_values(by='support', ascending=False)
```

```python
#2.2 Generate Association Rules (min_confidence = 0.5)
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets,
                          metric="confidence",
                          min_threshold=0.5)

rules_sorted = rules.sort_values(by='lift', ascending=False)

rules_sorted.head()
```

```python
#2.3 Top 6 Rules (by lift)
```

```python
top6 = rules_sorted.head(6)

top6_clean = top6.drop(columns=[
    'antecedent support',
    'consequent support',
    'conviction'
])

top6_clean[['antecedents', 'consequents',
            'support', 'confidence',
            'lift', 'leverage']]
```

```python
#2.4 Translate highest lift rule
```

```python
best_rule = rules_sorted.iloc[0]

print(f"If {list(best_rule['antecedents'])} are purchased, "
      f"then with confidence {round(best_rule['confidence']*100,2)}% "
      f"{list(best_rule['consequents'])} will also be purchased. "
      f"This rule has a lift ratio of {round(best_rule['lift'],2)}.")
```

```python
#3.1 Load and Create Binary Matrix
```

```python
books = pd.read_csv("CharlesBookClub.csv")

columns_to_drop = ['Seq#', 'ID#', 'Gender', 'M', 'R', 'F',
                   'FirstPurch', 'Related Purchase']

books_binary = books.drop(columns=columns_to_drop, errors='ignore')

books_binary = books_binary.applymap(lambda x: 1 if x > 0 else 0)

books_binary.head(10)
```

```python
#3.2 Apriori (min_support = 200 transactions)
```

```python
frequent_books = apriori(books_binary,
                         min_support=0.05,
                         use_colnames=True)

print("Number of frequent itemsets:",
      len(frequent_books))
```

```python
#3.3 Association Rules (min_confidence = 0.5)
```

```python
book_rules = association_rules(frequent_books,
                               metric="confidence",
                               min_threshold=0.5)

book_rules_sorted = book_rules.sort_values(by='lift',
                                           ascending=False)

top25 = book_rules_sorted.head(25)

top25[['antecedents', 'consequents',
       'support', 'confidence',
       'lift', 'leverage']]
```

```python
# 4.1 Rule with Highest Support
```

```python
#6.1 Create Synthetic Ratings (5000 ratings)
```

```python
np.random.seed(0)

ratings = pd.DataFrame({
    'userID': np.random.randint(0, 1000, 5000),
    'itemID': np.random.randint(0, 100, 5000),
    'rating': np.random.randint(1, 6, 5000)
})

ratings.head(10)
```

```python
#6.2 Convert to Surprise Format
```

```python
import pandas as pd
import numpy as np
from surprise import Dataset, Reader
from surprise.model_selection import train_test_split
from surprise import KNNBasic
from collections import defaultdict

np.random.seed(42)
num_users = 100
num_items = 50
num_ratings = 5000

ratings = pd.DataFrame({
    'userID': np.random.randint(1, num_users+1, size=num_ratings),
    'itemID': np.random.randint(1, num_items+1, size=num_ratings),
    'rating': np.random.randint(1, 6, size=num_ratings) 
})

reader = Reader(rating_scale=(1,5))
data = Dataset.load_from_df(ratings[['userID','itemID','rating']], reader)

trainset, testset = train_test_split(data, test_size=0.2)

sim_options = {'name': 'cosine', 'user_based': False}
algo = KNNBasic(sim_options=sim_options)
algo.fit(trainset)

predictions = algo.test(testset)

def get_top_n(predictions, n=5):
    top_n = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        top_n[uid].append((iid, est))
    for uid, user_ratings in top_n.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_n[uid] = user_ratings[:n]
    return top_n

top_n = get_top_n(predictions, n=5)

for uid in list(top_n.keys())[:5]:
    print(f"User {uid}: {top_n[uid]}")
```

```python
#6.3 Item-Based CF (Cosine Similarity)
```

```python
from surprise import KNNBasic

sim_options = {
    "name": "cosine",
    "user_based": False
}

algo = KNNBasic(sim_options=sim_options)
algo.fit(trainset)
```

```python
#6.4 Predict Missing Ratings & Recommend Items
```

```python
predictions = algo.test(testset)

from collections import defaultdict

top_recs = defaultdict(list)

for pred in predictions:
    top_recs[pred.uid].append((pred.iid, pred.est))


for user, items in list(top_recs.items())[:5]:
    items.sort(key=lambda x: x[1], reverse=True)
    print(f"User {user} recommended item {items[0][0]} "
          f"with predicted rating {round(items[0][1],2)}")
```
